```python
pdata.dropna(inplace=True)
pdata.reset_index(inplace=True)
pdata
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=pdata, x='Source')
plt.title('Distribution of Source')
plt.xlabel('Source')
plt.ylabel('Count')
plt.show()
```

```python
pdata['Duration'] = pdata['Duration'].str.replace('h', '*60').str.replace(' ', '+').str.replace('m', '*1').apply(eval)

plt.figure(figsize=(10, 6))
sns.histplot(data=pdata, x='Duration', bins=20, kde=True)
plt.title('Distribution of Duration')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```
